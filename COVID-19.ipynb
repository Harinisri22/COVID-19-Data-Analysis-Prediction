```python
X=covid_19.drop(['Date','Total Confirmed cases', 'Death','New cases', 'New deaths'],axis=1)
Y=covid_19['New recovered']
```

```python
print("Missing values in X before:", X.isnull().sum().sum())
print("Missing values in Y before:", Y.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
```

```python
mask = ~Y.isna()
X = X.loc[mask]
Y = Y.loc[mask]
```

```python
print("Missing values in X after:", X.isnull().sum().sum())
print("Missing values in Y after:", Y.isnull().sum())
```

```python
covid_19['Recovery_Class'] = covid_19['New recovered'].apply(lambda x: 1 if x > 0 else 0)
Y = covid_19['Recovery_Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
X_test.tail()
```

```python
Y_test.head()
```

```python
Y_train.tail()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)

print("Test Accuracy:", round(accuracy*100, 2), "%")
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
```

```python
print("\nClassification Report:\n", classification_report(Y_test, Y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = covid_19["Name of State / UT"]
y = covid_19["New cases"]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='darkblue', alpha=0.6)

plt.xlabel("States/UT")
plt.ylabel("New COVID-19 Cases")
plt.title("COVID-19 New Case Distribution by State")

plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
state_cases = covid_19.groupby("Name of State / UT")["New cases"].sum().reset_index()

plt.figure(figsize=(8, 5))
plt.plot(state_cases["Name of State / UT"], state_cases["New cases"], marker='s', color='red')

plt.xlabel("States/UT")
plt.ylabel("Total New Cases")
plt.title("COVID-19 Case Count by State")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x="Name of State / UT",
    y="New cases",
    hue="Name of State / UT",  # assign x variable as hue
    data=state_cases.sort_values("New cases", ascending=False),
    palette="plasma",
    legend=False   # avoid duplicate legend
)
plt.xticks(rotation=90)
plt.title("COVID-19 New Cases by State")
plt.show()
```
